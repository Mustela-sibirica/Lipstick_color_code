```python
# -*- coding: utf-8 -*-
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

name= []
Hex = []
R = []
G = []
B = []
with open(r'C:\Users\SUN\Desktop\v0.1.csv') as f:
    f_csv = csv.reader(f)
    headers = next(f_csv)
    for row in f_csv:
        code_list = list(row[1])
        for i in code_list:
            code_1 = int((code_list[1] + code_list[2]),16)
            code_2 = int((code_list[3] + code_list[4]),16)
            code_3 = int((code_list[5] + code_list[6]),16)
            item = [row[0], code_1, code_2, code_3]
        name.append(item)
        Hex.append(row[1])
        R.append(code_1)
        G.append(code_2)
        B.append(code_3)
fig = plt.figure()
ax = plt.subplot(111,projection='3d')
ax.scatter(R, G, B, c=Hex)
ax.set_xlabel("R")
ax.set_ylabel("G")
ax.set_zlabel("B")
plt.show()
```
